```python
#Ejercicio 15 ¿Como obtener los items que no son comunes en una serie A y serie B? (INCOMPLETO)
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
mapeo=ser1.map(ser2)
indices=np.isnan(mapeo)
resultado=mapeo[indices]
print(resultado)
```

```python
#Ejercicio 16 ¿Como obtener el numero de veces que se repite un valor en una serie?
ser = pd.Series(np.take(list('abcdefgh'), np.random.randint(8, size=30)))
df=pd.DataFrame(ser,columns=['letras'])
df['letras'].value_counts()
```

```python
#Ejercicio 17 ¿Como mantener los 2 valores mas repetidos de una serie, y a los demas valores cambiarles por 0 ?
np.random.RandomState(100)
ser = pd.Series(np.random.randint(1, 5, [12]))
df=pd.DataFrame(ser,columns=['values'])
top_valores=df['values'].value_counts().head(2)
indices=top_valores.index
mapeo=ser.map(top_valores)
indices_a_reemplazar=np.isnan(mapeo)
mapeo[indices_a_reemplazar]=0
mapeo
```

```python
#Ejercicio 20 ¿Como anadir series vertical u horizontalmente a un DataFrame?
ser1 = pd.Series(range(5))
ser2 = pd.Series(list('abcde'))
#Verticalmente
df_vertical=pd.DataFrame(ser1,columns=['Serie 1'])
df_vertical['Serie 2']=ser2
print(df_vertical)
print()
#Horizontalmente
df_horizontal=pd.DataFrame((ser1,ser2), index=['Serie 1','Serie 2'])
print(df_horizontal)
```

```python
# Ejercicio 21 ¿Obtener la media de una serie agrupada por otra serie?
frutas = pd.Series(np.random.choice(['manzana', 'banana', 'zanahoria'], 10))
pesos = pd.Series(np.linspace(1, 10, 10))
df=pd.DataFrame(frutas,columns=['frutas'])
df['pesos']=pesos
agrupacion=df.groupby('frutas').mean()
print(pesos.tolist())
print(frutas.tolist())
print(agrupacion)
```
